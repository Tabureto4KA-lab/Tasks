```python
import pandas as pd
data = pd.read_csv('data_csv.csv', sep=',')
data.head()
```

```python
fig = plt.figure('График с заданными параметрами', figsize=(6,6)) #верхние границы
plt.grid(True) #сетка
plt.title('Динамика цены золота на бирже', fontsize = 15) #Название и размер
plt.xlabel('t, день', fontsize = 12) #Подпись x
plt.ylabel('P, цена $', fontsize = 12)
plt.xticks(np.arange(start = 0, stop = len(data), step = 100), rotation = 17, size = 10)
# изменение шага меток x, их поворот и размер
plt.yticks(size = 10)
plt.plot(data['date'], 
        data['price'], 
        marker='.',
        color='crimson',
        markerfacecolor = 'black',
        markeredgecolor = 'black', 
        linewidth=2,
        markersize = 3)

plt.show()
```

```python
%matplotlib notebook
big = 13
med = 10
small = 2

params = {'figure.figsize': (5,5),
         'axes.titlesize': big,
         'axes.labelsize': med,
         'lines.linewidth': small,
         'lines.markersize': small*2,
         'xtick.labelsize': med,
         'ytick.labelsize': med,
         'lines.marker': '.',
         'lines.markerfacecolor': 'black',
         'lines.markeredgecolor': 'black',
         'axes.grid': True}

plt.rcParams.update(params) #применить параметры ко всем графикам


fig = plt.figure('График с заданными параметрами')
plt.title('Динамика цены золота на бирже')
plt.xlabel('t, день')
plt.ylabel('P, цена')
plt.plot(data['date'], data['price'], color = 'crimson')
plt.show()
```

```python
%matplotlib inline
x = np.linspace(-3,3,20)
fig = plt.figure(('3 графика на одном холсте'), figsize = (5,5))
plt.plot(x, x, label = 'Линейная')
plt.plot(x, x**2, label = 'Квадратическая')
plt.plot(x, x**3, label = 'Кубическая')
plt.xlabel('x label')
plt.ylabel('y label')
plt.title("Вывод трех графиков на одном полотне")
plt.legend(loc='upper center')

plt.show()
```

```python
fig = plt.figure(figsize=(7,7))
axes = fig.add_subplot(projection='3d')

X = np.arange(-10, 10, 0.05)
Y = np.arange(-10, 10, 0.05)
X, Y = np.meshgrid(X, Y)
Z = X**2 - Y**2

surf = axes.plot_surface(X, Y, Z, cmap = 'gist_rainbow')
axes.set_xlabel ('X')
axes.set_ylabel ('Y')
axes.set_zlabel ('Z')
axes.set_title ('Z = X**2 - Y**2')
#axes.view_init(45, 135)
plt.show()
```

```python
np.random.seed(2718281828)
fig = plt.figure('График разброса', figsize = (5,5))
x = np.random.randint(0, 100, 100)
y = np.random.randint(0, 100, 100)
s = np.random.randint(0, 500, 100)
c = np.random.randint(0, 500, 100)

plt.title('График разброса')
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(x, y, s, c)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(12, 6))
energy = [1813.70, 4206.14, 1023.10, 749.99]
name = ['Wnd', 'Hydropower', 'Solar', 'Other']

ax[1].bar(name, energy, color = 'teal', edgecolor = 'black')
ax[1].grid(alpha = 0.3, zorder = 1)
ax[1].set_title('Диаграмма производства возобновляемой \nелектроэнергии в 2021 году')
ax[1].set_ylabel('Twh')

ax[0].barh(name, energy, color = 'teal', edgecolor = 'black')
ax[0].grid(alpha = 0.3, zorder = 1)
ax[0].set_title('Диаграмма производства возобновляемой \nелектроэнергии в 2021 году')
ax[0].set_ylabel('Twh')

plt.show()
```
