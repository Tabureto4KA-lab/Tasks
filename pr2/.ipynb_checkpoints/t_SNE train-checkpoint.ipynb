```python
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn import preprocessing
import seaborn as sns
import umap

data = pd.read_csv('zoo.csv')
data
```

```python
D = data.drop(['class_type', 'animal_name'], axis = 1)

scaler = preprocessing.MinMaxScaler()
D = pd.DataFrame(scaler.fit_transform(D), columns = D.columns)
D
```

```python
T = TSNE(n_components=2, perplexity=5, random_state=123)
```

```python
TSNE_features = T.fit_transform(D)
```

```python
TSNE_features[1:4,:]
```

```python
DATA=D.copy()
DATA['x']=TSNE_features[:,0]
DATA['y']=TSNE_features[:,1]
```

```python
%matplotlib
fig = plt.figure()
sns.scatterplot(x='x',y='y',hue=data['class_type'],data=DATA,palette='bright')
plt.show()
```
